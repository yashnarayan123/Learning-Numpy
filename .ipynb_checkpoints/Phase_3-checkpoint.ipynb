# Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("====Swiggy Sales Data======")
print("Shape of data: ",sales_data.shape) 
print("Sales data for first 3 restaurents\n ",sales_data[:3])
```

```python
# yearly sales of swiggy
yearly_total = np.sum(sales_data[:,1:],axis = 0)
print("Yearly sales of Swiggy",yearly_total) 

# minimum sales per restaurent
min_sales = np.min(sales_data[:,1:],axis = 0)
print("Minimum sales per resturent: ",min_sales)

#maximum sales of the restaurent
max_sales = np.max(sales_data[:,1:],axis = 0)
print("Maximum sales per restaurent: ",max_sales)

# Average sales per restaurent 
avg_sales = np.mean(sales_data[:,1:],axis = 0)
print("Average sales per year",avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:,1:],axis = 1)
print(cumsum)
plt.Figure(figsize=(6,3))
plt.plot(np.mean(cumsum,axis=0))
plt.xlabel("years")
plt.ylabel("Sales")
plt.grid(True)
plt.title("Average cumalative sale of all restaurents per year")
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([2,3,4,5,10])

print("Sum of the vectors: ",vector1+vector2)
print("Multiplication of vectors",vector1*vector2)
print("Dot product of vectors:",np.dot(vector1,vector2))

print("Angle b/w the vectors: ",np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2))))
```
